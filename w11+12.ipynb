```python
states = pd.Series(["OH", "CA", "CA", "OH", "CA","WB","WB"])
years = pd.Series([2005,2005,2005,2005,2007,2005,2006])
population = pd.Series ([404,400,405,403,388,345,234])

df = pd.DataFrame({"state": states, "year":years, "population": population})
df
grouped = df.groupby([df["state"], df["year"]])
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y, label="Sine Wave")
plt.title("Sine Wave", fontsize=16, fontstyle='italic')

# Set the axis labels with custom font size and style
plt.xlabel("x-axis", fontsize=12, fontstyle='normal')
plt.ylabel("y-axis", fontsize=12, fontstyle='normal')

plt.show()
```

```python
x = np.linspace(-np.pi,np.pi,20)
y = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x,y, label="Cosine Wave", linestyle="--", marker = "o", markerfacecolor="red")
ax.set_title("Cosine Wave", fontsize=16, fontstyle='italic')

# Set the axis labels with custom font size and style
ax.set_xlabel("x-axis", fontsize=12, fontstyle='normal')
ax.set_ylabel("y-axis", fontsize=12, fontstyle='normal')
ax.legend(loc="upper left")
plt.show()
```

```python

x = np.linspace(-np.pi,np.pi,20)
y = np.cos(x)


peak_y = y[np.argmax(y)]
peak_x = x[np.argmax(y)]

fig, ax = plt.subplots()
ax.plot(x,y, label="Cosine Wave", linestyle="--", marker = "o", 
        markerfacecolor="black", color="blue")
ax.set_title("Cosine Wave", fontsize=16, fontstyle='italic')

#annotating the peak points in the plot
text = f'Peak: ({peak_x:.2f}, {peak_y:.2f})'
ax.annotate(text, xy=(peak_x, peak_y), xytext=(peak_x - 2, peak_y - 0.01),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=8, color='black')

# Set the axis labels with custom font size and style
ax.set_xlabel("x-axis", fontsize=12, fontstyle='normal')
ax.set_ylabel("y-axis", fontsize=12, fontstyle='normal')
ax.legend()
plt.show()
```
